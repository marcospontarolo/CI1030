```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import glob
```

```python
datasets = []
df_final = []
count = 1
```

```python
# Lendo todos os datasets em csv contidos na pasta de datasets
# Pegando um porcao de 50% de cada dataset do trafego normal
# Pegando uma porcao de 0,5% de cada dataset do trafego DDoS (de todos os tipos de ataque DDoS)
# Juntando em um Dataset final
for i in glob.glob("dataset/csv/01-12"+"/*.csv"):
    df = pd.read_csv(i)
    datasets = []
    df.drop('Unnamed: 0', axis=1, inplace=True)
    df.columns = df.columns.str.strip()
    print("\n----------------\n")
    print("\nDataset "+str(count))
    print("\nFile: "+str(i))
    print("\nTotal: "+str(df.shape))
    # um subset do nosso dataframe, somente com o trafego normal
    normal = df[df["Label"] == "BENIGN"]
    normal = normal.sample(frac=0.50)
    print("\nNormal: "+str(normal.shape))
    # um subset do nosso dataframe, somente com o trafego ddos
    ddos = df[df["Label"] != "BENIGN"]
    ddos = ddos.sample(frac=0.005)
    print("\nDDoS: "+str(ddos.shape))
    if count != 1:
        datasets.append(df_final)
    datasets.append(normal)
    datasets.append(ddos)
    count += 1
    df_final = pd.concat(datasets)
    print("\nFinal: "+str(df_final.shape))
```

```python
# Lista os tipos de ataques (ou trafego normal) do nosso dataset final
df_final.groupby(['Label']).size()
```

```python
# Salva o novo datasetcriado
df_final.to_csv("dataset/dataset_completo.csv", encoding='utf-8', index=False)
```
